```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the clean data
df = pd.read_csv('../data/clean.csv')
# Load original for categorical charts
df_raw = pd.read_csv('../data/train.csv')

print("Clean data shape:", df.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_raw['SalePrice'], bins=50, kde=True, color='steelblue')
plt.title('Distribution of House Sale Prices', fontsize=16)
plt.xlabel('Sale Price ($)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../images/01_price_distribution.png', dpi=150)
plt.show()
print("✅ Saved")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_raw, x='GrLivArea', y='SalePrice', alpha=0.5, color='steelblue')
plt.title('Living Area vs Sale Price', fontsize=16)
plt.xlabel('Above Ground Living Area (sqft)')
plt.ylabel('Sale Price ($)')
plt.tight_layout()
plt.savefig('../images/02_size_vs_price.png', dpi=150)
plt.show()
print("✅ Saved")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_raw, x='OverallQual', y='SalePrice', palette='Blues')
plt.title('Overall Quality vs Sale Price', fontsize=16)
plt.xlabel('Overall Quality (1-10)')
plt.ylabel('Sale Price ($)')
plt.tight_layout()
plt.savefig('../images/03_quality_vs_price.png', dpi=150)
plt.show()
print("✅ Saved")
```

```python
# Top 10 correlated features with SalePrice
top_features = df_raw.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)[1:11].index.tolist()
top_features.append('SalePrice')

plt.figure(figsize=(12, 8))
sns.heatmap(df_raw[top_features].corr(), annot=True, fmt='.2f', cmap='Blues')
plt.title('Correlation Heatmap — Top Features vs Sale Price', fontsize=16)
plt.tight_layout()
plt.savefig('../images/04_correlation_heatmap.png', dpi=150)
plt.show()
print("✅ Saved")
```
