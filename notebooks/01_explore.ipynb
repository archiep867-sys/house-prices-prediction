```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('../data/train.csv')

# First look
print("Shape:", df.shape)
df.head()
```

```python
# How many missing values in each column?
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# Basic statistics of all numeric columns
df.describe()
```

```python
# What's the average, min, max house price?
print("Average price: $", round(df['SalePrice'].mean()))
print("Cheapest house: $", df['SalePrice'].min())
print("Most expensive: $", df['SalePrice'].max())
```

```python
# Visualize house price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], bins=50, kde=True, color='steelblue')
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price ($)')
plt.ylabel('Count')
plt.savefig('../images/price_distribution.png')
plt.show()
```

```python
# Top 10 features most correlated with SalePrice
corr = df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
print(corr.head(11))  # top 10 + SalePrice itself
```
